```octave
N = 20;
steps = 2000;
dW = [zeros(1,N);mvnrnd(zeros(1,N),eye(N)./steps,steps)];
```

```octave
t = repmat([0:(1/steps):1]',1,N);
```

```octave
W = (dW'*triu(ones(steps+1)))';
```

```octave
plot(W)
xlim([0,steps])
```

```octave
X = W+t;
plot(X)
xlim([0,steps])
```

```octave
pi = exp(-W(end,:)-0.5);
pi
```

```octave
W_Q = W*diag(pi);
```

```octave
plot(W_Q)
xlim([0,steps])
```
